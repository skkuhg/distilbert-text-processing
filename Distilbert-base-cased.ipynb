```python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("fill-mask", model="distilbert/distilbert-base-cased")
```

```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForMaskedLM

tokenizer = AutoTokenizer.from_pretrained("distilbert/distilbert-base-cased")
model = AutoModelForMaskedLM.from_pretrained("distilbert/distilbert-base-cased")
```

```python
# Test the pipeline
result = pipe("Hello I'm a [MASK] model.")
for candidate in result:
  print(candidate)
```

```python
from transformers import DistilBertTokenizer, DistilBertModel
from google.colab import userdata

# Get the Hugging Face token from Colab secrets
hf_token = userdata.get('HF_TOKEN')

tokenizer = DistilBertTokenizer.from_pretrained('distilbert-base-uncased', token=hf_token)
model = DistilBertModel.from_pretrained("distilbert-base-uncased", token=hf_token)
text = "Replace me by any text you'd like."
encoded_input = tokenizer(text, return_tensors='pt')
output = model(**encoded_input)
```

```python
# Show the produced word embeddings
print(output.last_hidden_state)
```

```python
# Decode the input IDs to see the tokens
tokens = tokenizer.convert_ids_to_tokens(encoded_input['input_ids'][0])
print(tokens)
```
